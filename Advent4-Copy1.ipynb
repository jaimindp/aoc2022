```python
lines = open('input.txt')
lines = lines.readlines()
lines
```

```python
arr = [[[] for j in range(9)] for i in range(8)]
for i, line in enumerate(lines[:8]):
    print(line)
    for j in range(9):
        if line[j*4+1] != '':
            arr[i][j] = line[j*4+1]
    
arr
```

```python
moves = [i[:len(i)-1] for i in lines[10:]]
moves
```

```python
stack = [[] for i in range(9)]
for i, line in enumerate(lines[:8]):
    print(line)
    for j in range(9):
        if lines[i][j*4+1] != ' ':
            stack[j].append(lines[i][j*4+1])
    
stack
```

```python
import copy
```

```python
stacks = [i[::-1] for i in stack]
```

```python
# Stack for each line
# 'move 3 from 4 to 3'
stack_1 = copy.deepcopy(stack)
for i, move in enumerate(moves):
    one_move = move.split()
    
    _, n_s, _, f_s, _, d_s = one_move
    n, f, d = int(n_s), int(f_s), int(d_s)

    for _ in range(n):
        stacks[d - 1].append(stacks[f - 1].pop())
    
    one, two, three = int(one_move[1]), int(one_move[3]), int(one_move[5])
#     print('\n')
#     print(stack_1[three-1])
#     print(stack_1[two-1])
#     print(one, two, three)
#     print('\n', stack_1[two-1][:one], '\n')
    
    stack_1[three-1] = stack_1[two-1][:one] + stack_1[three-1]
    stack_1[two-1] = stack_1[two-1][one:]
    
#     print(stack_1[three-1])
#     print(stack_1[two-1])
    
# #     if i == 2:
# #         break
#     print('\nAfter\n')
#     print(stack_1)

# #     break
''.join(stack[-1] if stack else '' for stack in stacks)
# stack_1
```
