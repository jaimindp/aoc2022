```python
import requests
import json
import numpy as np
```

```python
file = open('input.txt')
lines = file.readlines()
```

```python
clean = [i.replace('\n', '') for i in lines]
    
clean = [int(i) if i != '' else 0 for i in clean]
```

```python
clean
```

```python
this_sum = 0
sums = []
for i in clean:
    if i == 0:
        sums.append(this_sum)
        this_sum = 0
    else:
        this_sum += i

sums.append(this_sum)
this_sum = 0    
```

```python
index_max = np.argmax(sums)
```

```python
index_max
```

```python
print(sums[index_max])
```

```python
for i in range(3):
    sums.pop(index_max)
    index_max = np.argmax(sums)
    print(sums[index_max])
```
