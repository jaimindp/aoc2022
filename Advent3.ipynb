```python
ord('a') - 96
ord('A') - 38

chr(97) - chr(99)
```

```python
this_dict = {}
for c in list(map(chr,range(ord('a'),ord('z')+1))):
#     print(c)
#     print(ord(c)-96)
#     print(ord(c.upper())-38)
    
    this_dict[c] = ord(c) - 96
    this_dict[c.upper()] = ord(c.upper()) - 38
```

```python
this_dict
```

```python
tot_sum = 0
for line in inputs:
      tot_sum += this_dict[list(set(line[:len(line)//2]).intersection(set(line[len(line)//2:])))[0] ]
tot_sum
```

```python
sum2 = 0
for i in range(0,len(inputs),3):
    sum2 += this_dict[list(set(inputs[i]).intersection(inputs[i+1]).intersection(inputs[i+2]))[0]]
sum2
```
