```python
tot = 0
for line in inputs:
    
    sp = line.split(',')
    x1, x2 = sp[0].split('-')
    y1, y2 = sp[1].split('-')
    x1 = int(x1)
    x2 = int(x2)
    y1 = int(y1)
    y2 = int(y2)

    if (x1 <= y1 and x2 >= y2):
        tot += 1
#         print(1, sp)
    elif (y1 <= x1 and y2 >= x2):
        tot += 1
#         print(2, sp)
    
tot
```
